```python
import numpy as np 
```

```python
def print_sudoku(arr): # arr is a 9x9 numpy array
    for i in range(9):
        for j in range(9):
            x = arr[i, j] if arr[i, j] != 0 else "."
            print(f" {x} ", end="")
            if j in [2, 5]:
                print("\u2551", end="")
        print("")
        if i in [2, 5]:
            print("\u2550"*9 + "\u256C" + "\u2550"*9 + "\u256C" + "\u2550"*9)
```

```python
with open ('sudokus.txt','r') as f:
    k=[]
    y=f.readlines()#reading all the lines in 'sudokus.txt' 
    for i in y: 
        i=i.rstrip()#stripping all the 'next line indicators'(/n) off the file by going line by line of the txt. file
        k.append(i) #storing the resulting output into an empty array 'k'
```

```python
n=[]
for m in k:
    if m != "": #since stripping all the new line indicators creates empty strings '' between the sudokus, so I stripped the empty strings off by appending everything in the array other than the empty strings
        n.append(m) 
n
```

```python
v=[]
for l in range(int(len(n)/9)):
    mm=n[l*9:l*9+9]#slicing the big array into subarrays by the number of lines in each sudoku, which is nine in this case, so each one of the subarrays represents a set of sudoku
    v.append(mm)
v
```

```python
bb=[]
for c in v:
    c=''.join(c)#using the join function to combine all the strings in each subarray
    c=c.replace(" ","")
    c=list(c)#using the list function to split up all the individual characters in the subarray in order to perform the reshape procedure later
    c=np.array(c)#converting the subarrays to numpy arrays
    c=c.reshape(9,9)#reshaping the subarrays to 9x9 (9 rows and each row contains 9 characters), since that's the requirement to utilize the given 'print_sudoku' funciton
    bb.append(c)
bb
```

```python
def checkrow(x):
    sett={'1','2','3','4','5','6','7','8','9'} #all the possible numbers in the sudoku
    setr=set(x) #since I'm going to put this in a for loop and use the x[] operation to get the specific row, so I just use set(x) to get the set of numbers in x
    setr=sett-setr #getting the remaining possible number that could be in that specific row
    return setr
```

```python
def checkcol(x,j):
    sett={'1','2','3','4','5','6','7','8','9'}
    setc=set(x[:,j]) #since I'm going to iterate through all the columns so I use x[:,j] to get the set of numbers in the jth column
    setc=sett-setc #getting the remaining possible number that could be in the jth column
    return setc
```

```python
def checkbox(x,i,j):
    sett={'1','2','3','4','5','6','7','8','9'}
    e=x[(i//3)*3:(i//3)*3+3,(j//3)*3:(j//3)*3+3].ravel()#using the integer division to slice the corresponding rows and columns for x[i,j]. They are both three in this case, since a box in a sudoku contains 9 characters, three rows and three columns, and by using the funciton '.ravel()' allows us to get the one dimension version of the 'box', which further allows us to get the set of numbers in that box
    setb=sett-set(e) #getting the remaining possible number that could be in the ith jth box
    return setb
```

```python
def stevendidit(x):
    while "0" in x: #while'0' in the sudoku, meaning as long as the sudoku isn't completely solved yet, we keep looping
        for i in range(len(x)): #getting the rows of the sudoku
            for j in range(len(x)): #getting the columns of the sudoku
                if x[i,j]=="0": #if the number in x[i,j] equals to '0', then we proceed to checking procedure
                    r=checkrow(x[i]) #checking the ith row
                    c=checkcol(x,j) #checking the jth column
                    b=checkbox(x,i,j) #checking the box of ith row and jth column
                    if len(r&c&b)==1: #if the intersection of the three contains only one character then we are certain that it has to be the right number of it
                        x[i,j]=(r&c&b).pop() #since the intersection only has one character, using the pop function will gurantee that we get the right number into the right spot assuming the codes above are correct
    return x
```

```python
for i in bb: #iterating through the array that contains all the subarrys that are in the format that could be used in the 'print_sudoku' function 
    print(print_sudoku(stevendidit(i))) #printing the resulting sudokus that are solved using the 'stevendidit' function
    #note I added another the print function to print it so the return None can seperate the sudokus
    
    
```
